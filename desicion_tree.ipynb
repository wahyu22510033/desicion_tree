# Decision Tree Studi Kasus ke 2

## Membaca database pada dataset file decisiontree_1.csv

```python
import pandas as pd
data = pd.read_csv('decisiontree_1.csv')
encoding = {"mesin" : {"bensin" : 0, "diesel" : 1}, 
            "penggerak" : {"depan" : 0, "belakang" : 1 }}
data.replace(encoding, inplace=True)
```

## Menampilkan Data Frame

```python
 data.head()
```

## Menampilkan Data Training

```python
X = data.drop(['ID','label'], axis =1)
Y = data['label']
```

## Membagi data training

```python
import sklearn.model_selection  as ms
X_train, X_test, Y_train, Y_test = ms.train_test_split(X,Y, test_size =0.3 ) 
```

## Import Library Decision Tree

```python
import sklearn.tree as tree
model = tree.DecisionTreeClassifier(criterion='gini', max_depth=3)
model.fit(X_train, Y_train)
```

## Pengujian Dataset

```python
y_prediksi = model.predict(X_test)
y_prediksi
```

## Pengujian Tingkat Akurasi

```python
import sklearn.metrics as met
print (met.accuracy_score(Y_test, y_prediksi))
```

# Visualisasi
## Visualisasi Gambar

```python
import matplotlib.pyplot as plt
from sklearn import tree

plt.subplots(figsize=(10,10))
tree.plot_tree(model, fontsize=10)
plt.show()
```
